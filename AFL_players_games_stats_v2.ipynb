 ### The datasets to be loaded

  * 'AFL_data/stats.csv'
  * 'AFL_data/players.csv'
  * 'AFL_data/games.csv'

*NB we aim to load these into a single dataframe by merging the three data sets.*


```python
# Reading datafiles into DataFrames and displaying them to do a first visual inspection

import pandas as pd
players = pd.read_csv("AFL_data/players.csv")
games = pd.read_csv("AFL_data/games.csv")
stats = pd.read_csv("AFL_data/stats.csv")

# Display the first few (and last) rows of each DataFrame

print("\nPlayers dataFrame: . . . ")
display(players)
print("\nGames dataFrame: . . . ")
display(games)
print("\nStats dataFrame: . . . ")
display(stats)
```


This approach sets up the foundation for further analysis, giving us a quick overview of the datasets before proceeding to detailed operations like cleaning, transformation, or visualization. It's a helpful first step in understanding the structure and quality of the data at hand. 


### Structure & organization of the datasets

We now retrieve and display the column headers (names) of the three datasets, which in the final print() statement combines all these lists into an organized, readable format, clearly labeling and separating the column headers for each dataset. This step is crucial for gaining an understanding of the data structure before performing analyses or transformations.

```python
# Curious about the column header names, esp. the full set from the 'stats' datafile

column_headers_players = list(players.columns)
column_headers_games = list(games.columns)
column_headers_stats = list(stats.columns)
print('PLAYER Column Headers: ...\n\n', column_headers_players, '\n\nGAMES Column Headers: ...\n\n', column_headers_games, '\n\nSTATS Column Headers: ...\n\n', column_headers_stats,)
```

### Merging the datasets

The code snippets in Cells [4] and [6] aim to combine multiple datasets into a single unified DataFrame, allowing for integrated analysis of AFL game statistics, player performance metrics, and associated game details.

```python
# Merging the 'players' and 'stats' DataFrames, using 'playerId' using an inner join.

merged_players_stats = pd.merge(players, stats, on= 'playerId', how= 'inner')
display(merged_players_stats)
```

```python
# Curious about the merged column header names.

column_headers_merged_players_stats = list(merged_players_stats.columns)
print('Merged df PLAYER-STATS Column Headers\n', column_headers_merged_players_stats)
```

```python
# Merging the resulting DataFrame, 'column_headers_merged_players_stats', with the 'games' DataFrame on 'gameId' using an inner join.

Merged_df = pd.merge(merged_players_stats, games, on= 'gameId', how= 'inner')
display(Merged_df)
```

Note the use of an 'inner join' method, which is used to combine two tables or datasets based on a common key column. The key column contains shared values between the two datasets, ensuring only the rows with matching values in the specified key column are included in the resulting merged dataset. In the first instance we joined thw two datasets based on the 'playerId' column, and in the second instance on the 'gameId' column.

### Transforming datatypes in the merged dataset, pre-analysis

The follwing snippet intends to clean and prepare the dataset for further analysis by addressing a specific issue identified, above. That is transforming 'object' datatype (typically a string) into the datatype of a date column, proper, in the Merged_df DataFrame.

```python
# Noting that the 'date' and 'dob' columns are datatype 'object',they may need to be transformed into 'datetime'

Merged_df['date'] = pd.to_datetime(Merged_df['date'])
Merged_df['dob'] = pd.to_datetime(Merged_df['dob'])

print("\nAfter conversion to DATETIME: \n")
print(Merged_df.info())
```

## Examining the performance of a player

Let us now look at a performing some data analysis.

For a particular player - say "Franklin, Lance" - let us enumerate and visualise his accumlation of goals over time. The visualisation will comprise a plot with the x-axis showing the number of games played and the y-axis showing the number of goals accumulated. Thus we expect to see a line that monotonically increases, but 'in what way?' will depend on the player's career, e.g. positions played, game strategies chosen, and personal health.

The following code snippets and accompanying markdown offer a comprehensive exploration of Lance Franklin's AFL performance, focusing on his goal-scoring progression across his career. The process begins by filtering the merged dataset to extract Franklin's data, effectively narrowing the scope to his personal statistics. This filtered data is saved in a dedicated CSV file, ensuring accessibility and facilitating further analysis beyond the current environment.

```python
# To start i'll make a filter for "Franklin, Lance"

player_name= "Franklin, Lance"
player_data= Merged_df[Merged_df['displayName_x'] == player_name]
display(player_data)

Merged_df.to_csv('LanceFranklin sport stats.csv', index=False)
```

Next, the timeline of Franklin's games and goal statistics is prepared. By sorting the data in chronological order, the analysis ensures that subsequent visualizations maintain temporal integrity. A pivotal step involves calculating Franklin's cumulative goals, creating a metric that tracks his goal-scoring accumulation over time—a rich indicator of his contributions on the field. These cumulative goals are then visualized using Matplotlib in two distinct plots: one mapping the progression of goals against the game dates, and the other charting the accumulation of goals against the number of games played. Both visualizations provide a clear picture of Franklin's career trajectory, offering insights into his consistency, peaks, and potential influencing factors such as his health, team strategies, or positional changes.

```python
# Selecting columns related to the selected player's GOALS and game DATES.

player_goals= player_data[['displayName_x','date', 'gameNumber', 'Goals']]

# Ensuring date data (in datetime format) is sorted ascending.

player_data = player_data.sort_values(by= 'date', ascending= True)

print("Timeline of selected player's goals\n\n", player_goals)
```

```python
# Calculating cumulutave sum of GOALS

player_data['CumulativeGoals'] = player_data['Goals'].cumsum()
player_goals= player_data[['displayName_x','date', 'gameNumber', 'Goals', 'CumulativeGoals']]

print("Timeline of selected player's GOALS by DATE \n\n", player_goals)
```

```python
# Plotting the Accumulation of GOALS by DATE.

import matplotlib.pyplot as plt 

player_data.plot(x= 'date', y='CumulativeGoals', kind= 'line', marker='o', linestyle='-', color='b', figsize=(10, 6), title=f'FIG.1: Accumulation of Goals Over TIME (YEAR) for {player_name}')
plt.xlabel('date')
plt.ylabel('Number of Goals Accumulated')
plt.grid(True)
plt.show()
```

**NOTE:** Gaps in between the series of datapoints, evident at the ends of years and starts of following years, are likely to represent the breaks between AFL seasons. That is, AFL seasons generally start in March and finish in September, each year.

```python
# Calculating cumulutave sum of GAMES

player_data['Number_of_Games'] = range(1, len(player_data) +1)
player_goals_v2 = player_data[['displayName_x','date', 'Number_of_Games', 'Goals', 'CumulativeGoals']]

print("Timeline of selected player's GOALS by NUMBER OF GAMES \n\n", player_goals_v2)
```

```python
# Plotting the Accumulation of GOALS by NUMBER OF GAMES.

player_data.plot(x= 'Number_of_Games', y= 'CumulativeGoals', kind='line', marker='o', linestyle='-', color='b', figsize=(10, 6), title=f'FIG.2: Accumulation of Goals Over Number of Games Played for {player_name}')
plt.xlabel('Number of Games Played')
plt.ylabel('Number of Goals Accumulated')
plt.grid(True)
plt.show()
```

Finally, the analysis concludes with descriptive statistics for the player's filtered data, adding depth and quantitative precision to the narrative.

```python
display(player_goals_v2.describe())
```

## Examining the performance of multiple players, simultaneously

In the one chart, let us now plot the lines for the following players:

* "Franklin, Lance"
* "Papley, Tom"
* "Mumford, Shane"
* "Hooker, Cale".

*The plot will show each player in a different colour so they can be easily distinguished. They are also referenced in a Legend.*


```python
# Creating masks for each player
mask_franklin = Merged_df['displayName_x'] ==  'Franklin, Lance'
mask_papley = Merged_df['displayName_x'] ==  'Papley, Tom'
mask_mumford = Merged_df['displayName_x'] ==  'Mumford, Shane'
mask_hooker = Merged_df['displayName_x'] ==  'Hooker, Cale'

# Creating one combined mask to include each player, with the OR operator
combined_mask = mask_franklin | mask_papley | mask_mumford | mask_hooker

# Applying the combined mask to filter the data for the selected players
filtered_data = Merged_df[combined_mask]

# Resetting the Index, by dropping it (to avoid later comparisons against 'zero' values)
# filtered_data_IndexReset = filtered_data.reset_index(drop=True)  - - - - -> doesn't work


# Viewing the date for its Attributes, Staistics, and Display
print('\n Summary of data ATTRIBUTES\n') 
print(filtered_data.info()) 
print('\n Summary of data STATISTICS\n')
print(filtered_data.describe()) 
display(filtered_data.head(), filtered_data.tail())

# For 'posterity' creating a .csv file from it for download
filtered_data.to_csv('filtered_data.csv', index=False)
```

```python
# Readying the plot

# Making list of Players
players =["Franklin, Lance", "Papley, Tom", "Mumford, Shane", "Hooker, Cale"]

# Creating a plot for the ACCUMULATION of GOALS by the NUMBER of GAMES PLAYED
plt.figure(figsize=(10, 6))

# Defining colours for players
colors = {'Franklin, Lance': 'blue', 'Papley, Tom': 'green', 'Mumford, Shane': 'red', 'Hooker, Cale': 'black'}

for player_name in players:
    # Filtering the data for our target Players.
    player_data = filtered_data[filtered_data['displayName_x'] == player_name]

    # Ensuring the date values are in ascending order.
    player_data = player_data.sort_values(by= 'date')

    # Calculating the cumulative sum of goals.
    player_data['CumulativeGoals'] = player_data['Goals'].cumsum()

    # Adding a column for the number of games plauyed.
    player_data['Number_of_Games'] = range(1, len(player_data) + 1)
    
    # Plotting the data for the players
    plt.plot(player_data['Number_of_Games'], player_data['CumulativeGoals'], linestyle= '-', label= player_name)

# Doing the titles and labels.
plt.title('FIG.3: Accumulation of Goals by NUMBER OF GAMES played by selected players')
plt.xlabel('Accumulated number of GAMES PLAYED')
plt.ylabel('Accumulated number of GOALS SCORED')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Supplementary plot, to show Accumulation of Goals Scored by DATE / YEAR / TIME

# Making list of Players
players =["Franklin, Lance", "Papley, Tom", "Mumford, Shane", "Hooker, Cale"]

# Creating a plot for the ACCUMULATION of GOALS by TIME/YEAR
plt.figure(figsize=(10, 6))

# Defining colours for players
colors = {'Franklin, Lance': 'blue', 'Papley, Tom': 'green', 'Mumford, Shane': 'red', 'Hooker, Cale': 'black'}

for player_name in players:
    # Filtering the data for our target Players.
    player_data = filtered_data[filtered_data['displayName_x'] == player_name]

    # Ensuring the date values are in ascending order.
    player_data = player_data.sort_values(by= 'date')

    # Calculating the cumulative sum of goals.
    player_data['CumulativeGoals'] = player_data['Goals'].cumsum()

    # Adding a column for the number of games played.
    player_data['Number_of_Games'] = range(1, len(player_data) + 1)
    
    # Plotting the data for the players
    plt.plot(player_data['date'], player_data['CumulativeGoals'], linestyle= '-', label= player_name)

# Doing the titles and labels.
plt.title('FIG.4: Accumulation of Goals by Number of Games played by TIME (YEAR) by selected players')
plt.xlabel('date')
plt.ylabel('Accumulated number of GOALS SCORED')
plt.grid(True)
plt.legend()
plt.show()
```

## Linear Regression as a means to predict goals scored over time

This plot will now show and compare two players, "Franklin, Lance" and "Hooker, Cale", but will also include the linear regression line for each. In other words. as well as showing their actual cummulative goals over time, we will plot their predicted goals over time where the prediction is done via Linear Regression. Again, different colours for each line (and player) are used and a Legend is included.


```python
# Needing to use a 'polyfit' function from numpy library for the regression line
import numpy as np

# Making list of Players
players =["Franklin, Lance", "Hooker, Cale"]

# Creating a plot for the ACCUMULATION of GOALS by the NUMBER of GAMES PLAYED
plt.figure(figsize=(10, 6))

# Defining colours for players
colors = {'Franklin, Lance': 'blue', 'Hooker, Cale': 'purple'}

for player_name in players:
    # Filtering the data for our target Players.
    player_data = filtered_data[filtered_data['displayName_x'] == player_name]

    # Ensuring the date values are in ascending order.
    player_data = player_data.sort_values(by= 'date')

    # Calculating the cumulative sum of goals.
    player_data['CumulativeGoals'] = player_data['Goals'].cumsum()


    # player_data['CumulativeGoals'] = player_data['CumulativeGoals'].reset_index(drop= True)


   # Adding a column for the number of games played.
    player_data['Number_of_Games'] = range(1, len(player_data) + 1)


    # Performing Linear Regression
    X = player_data['Number_of_Games'].values.reshape(-1,1)
    Y = player_data['CumulativeGoals'].values
    m, b = np.polyfit(player_data['Number_of_Games'], player_data['CumulativeGoals'], 1)
    player_data['PredictedGoals'] = m * player_data['Number_of_Games'] + b


    # Plotting the data for the players
    plt.plot(player_data['Number_of_Games'], player_data['CumulativeGoals'], linestyle= '-', label= player_name)

    # Plotting the predicted cumulative goals (regression)
    plt.plot(player_data['Number_of_Games'], player_data['PredictedGoals'], linestyle= '--', label= player_name)

# Doing the titles and labels.
plt.title('FIG.5: Accumulation of Goals by Number of Games played by selected players (ACTUAL & PREDICTED)')
plt.xlabel('Accumulated number of GAMES PLAYED')
plt.ylabel('Accumulated number of GOALS SCORED')
plt.grid(True)
plt.legend() 
plt.show()
```

### **INSIGHT**

- The regression line (Linear) shows a close correspondence against the dataplot of Lance FRANKLIN's accumulated goals scored across games played. Hence, we can expect good predictive value (for y) of this regression line from the accumulated number of games played (per x-values).

- On the other hand, the dataplot of Cale HOOKER's accumulated goals scored across games played, shows wide variability and divergences against the (Linear) regression line plotted. Thus, while it may uphold the general trend of accumulated goals scored across games played, its predictive value (from x-values) is decidedly questionable. The variability in the dataplot here is related to an earlier discussion that game strategy or position play of Cale HOOKER possibly changed across the course of his AFL career. It is more suitable, then, to apply a Moving Average regression (contra a Linear regression) which helps capture fluctuations, short-term spikes, etc., in the data, thereby better capturing the underlying trend, and potentially better predictive value (for y) from the accumulated number of games played (per x-values).
